```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('jpg')
```

```python
cat = pd.read_csv('../../hsc-diffuse-galaxies/data/final-sample/hsc-lsbg-final-cat.csv')
cat.drop('id', axis=1, inplace=True)
cat.rename(columns={'cat-id':'id'}, inplace=True)
cat.set_index('id', drop=True, inplace=True)
galex = pd.read_csv('../../hsc-diffuse-galaxies/paper/table-1-data/tab-1-galex.csv')
galex['id'] = [int(i.split('-')[1]) for i in galex['id']]
galex.set_index('id', drop=True, inplace=True)
ids = np.array([662, 522, 605, 40, 614, 741, 745, 515, 540, 497])
```

```python
fig, ax = plt.subplots(2, 1, figsize=(5,7))
fig.subplots_adjust(hspace=0.33)

ax[0].hist(galex.loc[ids].nuv, bins='auto', range=[19.5, 21.5])
ax[1].hist(cat.loc[ids].mu_0_forced_g, bins='auto')

ax[0].set_xlabel('NUV [mag]', fontsize=17)
ax[1].set_xlabel(r'$\mu_0(g)\ \mathrm{[mag\ arcsec^{-2}]}$', fontsize=17)
ax[0].set_xlim(19.4, 21.6)
ax[1].set_xlim(22.9, 25.6)

ax[0].set_title('Target Distributions')

for i in range(2):
    ax[i].set_ylim(0, 4.5)
    ax[i].set_yticks(np.arange(0,5,1))
    ax[i].set_ylabel('# of Galaxies', fontsize=18)
    
#fig.savefig('/Users/protostar/Dropbox/Apps/ShareLaTeX/Gemini-Proposal-LSB-2018A/target-dists.pdf')
```
